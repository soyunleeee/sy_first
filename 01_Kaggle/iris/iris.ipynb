3) model 선정
4) model.fit(문제지8, 답안지8)
5) 컴퓨터답2 = model.predict(문제지2)
6) accuracy_score(답안지2, 컴퓨터답2)

```python
정답지 = df["target"]
문제지 = df.drop("target", axis=1)
# df2 = df[~ "target"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split
```

```python
문제지8, 문제지2, 정답지8, 정답지2 = train_test_split(문제지, 정답지, train_size=0.8, random_state=1414)
```

```python
문제지8.shape , 정답지8.shape,  문제지2.shape, 정답지2.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
model = RandomForestClassifier()
def train_fit_score(문제지, 정답지, model=model, random_state=None):
    
    문제지8, 문제지2, 정답지8, 정답지2 = train_test_split(문제지, 정답지, train_size=0.8, random_state=1414)
    model.fit(문제지8, 정답지8)
    컴퓨터답2 = model.predict(문제지2)
    score = accuracy_score(정답지2, 컴퓨터답2)
    print("1차점수 : accuracy:", score)
    
train_fit_score(문제지, 정답지, model, random_state=1414)
```

# kfold

```python
from sklearn.model_selection import KFold, StratifiedKFold
```

```python
kfold = KFold(n_splits=3 , random_state=1414, shuffle=True)
```

```python
total_score = []
for train_index, test_index in kfold.split(문제지, 정답지):
    문제지8, 문제지2 = 문제지.iloc[train_index], 문제지.iloc[test_index]
    정답지8, 정답지2 = 정답지.iloc[train_index], 정답지.iloc[test_index]
   #train_test_split(문제지,정답지, test_sizse=0.2, random_state=1414, shuffle=False) ----
    model.fit(문제지8, 정답지8)
    컴퓨터답2 = model.predict(문제지2)
    score = accuracy_score(정답지2, 컴퓨터답2)
    print("fold accuracy:", score)  
    total_score.append(score)
    ####return np.array(total_score)
    
print("평균", np.array(total_score).mean()) 
```
