```python
model = RandomForestClassifier()
def split_fit_score(문제지, 정답지, model=model, random_state=None):
    문제지8, 문제지2, 정답지8, 정답지2 = train_test_split(문제지, 정답지 ,test_size=0.8 , random_state=1414)
    model.fit(문제지8, 정답지8)
    컴퓨터답2 = model.predict(문제지2)
    score = f1_score(정답지2, 컴퓨터답2)
    print("1차점수", score)
    
split_fit_score(문제지, 정답지, model)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
y = df["outcome"]
X = df.drop("outcome", axis=1)
```

```python
X.columns
```

```python
X[X.columns] = scale.fit_transform(X[X.columns])
X.head()
```

```python
df2 = pd.concat([X, y], axis=1)
df2.head()
```

```python
from imblearn.over_sampling import SMOTE
y = df2["outcome"]
X = df2.drop("outcome",axis=1)

# ========================== SMOTE 오버샘플링 =============================
smote = SMOTE(random_state=0)
X_smote, y_smote = smote.fit_resample(X, y)
print(X_smote.shape, y_smote.shape)
```

```python
df_smote = pd.concat([X_smote, y_smote], axis=1)
df_smote.head()
```
