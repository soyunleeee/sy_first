# modeling

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(penalty='l2', max_iter=100)

lr.fit(X, y)

y_pred      = lr.predict(X_test)
probas_pred = lr.predict_proba(X_test)
```

```python
probas_pred.shape
```

# metrics

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score , roc_auc_score

acc = accuracy_score(y_test, y_pred)
pre = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1  = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, probas_pred[:, -1])

acc, pre, rec, f1, auc
```

# plot metrics

```python
from sklearn.metrics import precision_recall_curve, roc_curve

precisions, recalls, ths = precision_recall_curve(y_test, probas_pred[:, -1])
f1 = 2 * (precisions * recalls) / (precisions + recalls)

f, axs = plt.subplots(1, 2, figsize=(14, 6))

sns.lineplot(x=ths, y=precisions[:ths.shape[0]], ax=axs[0])
sns.lineplot(x=ths, y=recalls[:ths.shape[0]], ax=axs[0])
sns.lineplot(x=ths, y=f1[:ths.shape[0]], ax=axs[0])

fpr, tpr, ths = roc_curve(y_test, probas_pred[:, -1])

sns.lineplot(x=fpr, y=tpr, ax=axs[1])
sns.lineplot(x=[0, 1], y=[0, 1], ax=axs[1])
```
