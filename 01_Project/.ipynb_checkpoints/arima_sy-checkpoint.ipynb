# ARIMA

```python
# 종가 로그 변환
df['log_price'] = np.log(df['Close']) #Firstly we should take the logarithmic return from prices
```

```python
# 차분해서 전일 대비 변화율
df['pct_change'] = df['log_price'].diff() #after that let's take difference
```

```python
# Volatility(변동성) 계산
df['stdev'] = df['pct_change'].rolling(window=30, center=False).std()
df['Volatility'] = df['stdev'] * (252**0.5) # Annualize.
```

```python
# Volatility(변동성) 그래프 보기
plt.figure(figsize=(16,6))
df['Volatility'].plot()
plt.title("Rolling Volatility With 30 Time Periods By Annualized Standard Deviation")
plt.show()
```

```python
# nan 값 삭제
df = df.dropna()
```

```python
df.shape
```

```python
# 왜 100을 곱했지??
vol = df["Volatility"] * 100
```

```python
df.head()
```

```python
close_df = df[['Date','Volatility']]
close_df = close_df.sort_values('Date')
close_df = close_df.set_index('Date')
```

```python
close_df.head()
```

```python
close_df.plot()
```

```python
plot_acf(close_df)
plot_pacf(close_df)   
plt.show()
```

# 차분해서 정상성 확인

```python
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

diff_1 = close_df.diff(periods=1).iloc[1:]
diff_1.plot()
plt.xticks(rotation =90)
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(close_df, order=(1,1,1))

arima_res = model.fit()
# (trend='c', full_output=True, disp=1)
print(arima_res.summary())
```

```python
pig, ax = plt.subplots(figsize=(15,5))
arima_res.plot_predict(ax=ax)
plt.show()
```
